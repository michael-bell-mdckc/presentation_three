```python
#init lists for democrat years & republican years
dy = []
ry = []

for row in potusdf.itertuples():
    if row.potus == "Democrat" and row.year not in dy:
        dy.append(row.year) 
    elif row.potus == "Republican" and row.year not in ry:
        ry.append(row.year)
print(dy)
print(ry)
```

```python
# set up matrix for win/loss years for democrat or republican Pres
rwin = .1000
rlose = .0010
dwin = .0100
dlose = .0001

#create copy to write changes to
djdf1 = djdf.copy()

k = 2000 # to index for columm value
for i in range(2,22):  #column index for 2000 to 2019
    for j in range(17): #Row index
        if djdf.iat[j,i] > 0: # check if up for the year (win)
            if k in ry:
                djdf1.iat[j,i] = rwin # set as republican up
            else:
                djdf1.iat[j,i] = dwin # set as democrat up
        elif djdf.iat[j,i] < 0: # if down for the year (loss)
            if k in ry:
                djdf1.iat[j,i] = rlose # set as republican down
            else:
                djdf1.iat[j,i] = dlose # set as democrat down
    k+=1 # index year after each column calculation
    
djdf1
```
